```python
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
```

```python
img = Image.open("Images/sunset.jpeg")
plt.imshow(img)
image_array = np.array(img)
```

```python

def rotate_image_rgb(img, angle_deg):
    angle_rad = np.deg2rad(angle_deg)

    # Rotation Transofrmation matrix
    R = np.array([
        [np.cos(angle_rad), -np.sin(angle_rad)],
        [np.sin(angle_rad),  np.cos(angle_rad)]
    ])

    
    h, w, c = img.shape
    center = np.array([w / 2, h / 2])
    # create zero array with same size of original image
    rotated = np.zeros_like(img)

    for i in range(h):
        for j in range(w):
            # Calculate (x, y) coordinate relative to center
            coord = np.array([j, i]) - center

            # Rotate coordinate with using Rotation Matrix
            new_coord = R @ coord + center
            x_new, y_new = new_coord.round().astype(int)

            if 0 <= x_new < w and 0 <= y_new < h:
                rotated[y_new, x_new] = img[i, j]

    return rotated
```

```python
rotated_image = rotate_image_rgb(image_array , 45 )
plt.imshow(rotated_image)
```
